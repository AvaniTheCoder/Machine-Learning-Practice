```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Creating a sample health data Frame

data = {
    'Patient_id':[1,2,3,4,5],
    'Age':[25,32,45,28,60],
    'Gender':['Male','Female','Male','Male','Female'],
    'Blood_Pressure':[120/80,130/85,140/90,125/82,135/88],
    'BMI':[22.5,28.3,24.7,23.0,29.8],
    'Cholestrol_Level':['Normal','High','Normal','Low','High'],
    'Smoker':[True,False,False,True,False],
    'Exercise_hours':[3,2,1,4,0]
}

health_df = pd.DataFrame(data)
var=type(health_df['Blood_Pressure'])
print(var)
```

```python
#Different graphs

#1-Histogram of age
plt.figure(figsize=(8,6))
sns.histplot(health_df['Age'],bins=10,kde=True) #kde gets a waing line on graph
plt.title('Distribution of age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
#2-Gender pie chart
gender_counts=health_df['Gender'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(gender_counts,labels=gender_counts.index,autopct='%1.1f%%',startangle=90,colors=['pink','green'])
plt.title('Gender Distribution')
plt.show()
```

```python
#3-Blood pressure boxplot
plt.figure(figsize=(8,6))
sns.boxplot(x='Blood_Pressure',data=health_df)
plt.title('Bllod pressure distribution')
plt.xlabel('Blood pressure')
plt.show()
```

```python
#4-Bmi vs cholestrol level
plt.figure(figsize=(8,6))
sns.scatterplot(x='BMI',y='Cholestrol_Level',data=health_df,hue='Cholestrol_Level')
plt.title('Bmi vs Cholestrol')
plt.xlabel('BMI')
plt.ylabel('Cholestrol level')
plt.legend(title='Cholestrol level')
plt.show()
```

```python
#4-Bmi vs cholestrol level
plt.figure(figsize=(8,6))
sns.barplot(x='Cholestrol_Level',y='BMI',data=health_df,hue='Cholestrol_Level')
plt.title('Bmi vs Cholestrol')
plt.xlabel('Cholestrol level')
plt.ylabel('BMI')
plt.legend(title='Cholestrol level')
plt.show()
```

```python
#5-Bar chart of smokers vs non-smokers
smoker_counts=health_df['Smoker'].value_counts()
plt.figure(figsize=(6,6))
sns.barplot(x=smoker_counts.index,y=smoker_counts.values)
plt.title('Smoker vs Nonsmokers')
plt.xlabel('Smoker')
plt.ylabel('Count')
plt.show()
```

```python
#Next question
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random

#Creating a samle sales data Frame

data = {
    'Order_id':[1001,1002,1003,1004,1005],
    'Product_Name':['Product A','Product B','Product C','Product A','Product D'],
    'Quantity':[10,5,8,12,6],
    'Price':[15.99,10.49,22.9,15.99,30.99],
    'Discount':[0.05,0.1,0.07,0.05,0.15],
    'Customer_Id':['Cust001','Cust002','Cust003','Cust001','Cust004'],
    'Order_Date':pd.to_datetime(['2023-01-05','2023-02-10','2023-03-15','2023-01-20','2023-04-25'])
}

sales_df = pd.DataFrame(data)
```

```python
#1-Bar chart of product sales quantity
product_sales=sales_df.groupby('Product_Name')['Quantity'].sum().reset_index()
plt.figure(figsize=(10,6))
sns.barplot(x='Product_Name',y='Quantity',data=product_sales,hue='Quantity')
plt.title('Product sales quantity')
plt.xlabel('Product Name')
plt.ylabel('Quantity sold')
plt.xticks(rotation=45)
plt.show()
```

```python
#1-BLine plot of Total sales Over time
sales_df['Month']=sales_df['Order_Date'].dt.month
monthly_sales=sales_df.groupby('Month')['Price'].sum().reset_index()
plt.figure(figsize=(8,6))
sns.lineplot(x='Month',y='Price',data=monthly_sales,)
plt.title('Total sales over Time')
plt.xlabel('Month')
plt.ylabel('Total sales price')
plt.xticks(range(1,13))
plt.grid()
plt.show()
```

```python
#3-Pie chart of customer distribution
customer_counts=sales_df['Customer_Id'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(customer_counts,labels=customer_counts.index,autopct='%1.1f%%',startangle=90)
plt.title('Customer Distribution')
plt.show()
```

```python
#4-Box plot of price by product
plt.figure(figsize=(8,6))
sns.boxplot(x='Product_Name',y='Price',data=sales_df)
plt.title('Price Distribution by product')
plt.xlabel('Product name')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show()
```

```python
#5- Historogram of discounts
plt.figure(figsize=(8,6))
sns.histplot(sales_df['Discount'],bins=10,kde=True) #kde gets a waing line on graph
plt.title('Distribution of Discounts')
plt.xlabel('Discount')
plt.ylabel('Frequency')
plt.show()
```

```python
#scatter plot of price vs Quantity
plt.figure(figsize=(8,6))
sns.scatterplot(x='Price',y='Quantity',data=sales_df,hue='Product_Name')
plt.title('Price vs Quantity')
plt.xlabel('Price')
plt.ylabel('Quantity')
plt.legend(title='Product Name')
plt.show()
```

```python
#Assignment of shipping products..............!
import pandas as pd
import numpy as np
from random import randint
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    "Shipment_id":[101,102,103,104,105,106,107,108,109,110],
    "Product_name":["blue saree","red dress","orange saree","green dress","purple saree","white kurta","black dress","pink saree","heavy dress","chiffon saree"],
    "Weight(kg)":[1,2,1,2,0.5,0.8,0.95,1.5,1,2.1],
    "Shipping_cost":[20,24,20,24,10,16,18,15,20,25],
    "Shipping_date":pd.to_datetime(['2023-01-05','2023-02-10','2023-03-15','2023-04-20','2023-05-25','2023-06-05','2023-07-10','2023-08-15','2023-09-20','2023-10-25']),
    "Destination":["Mumbai","Kolkata","Bihar","Lucknow","Delhi","Jaipur","Mumbai","Delhi","Lucknow","Goa"],
    "Shipping_status":["shipped","shipped","delivered","shipped","delivered","delivered","shipped","shipped","shipped","shipped"],
    "Priority_shipping":[False,False,True,True,True,True,False,False,False,False],
    "Delivery_time":[3,4,3,5,3,7,4,3,5,5]
}
df=pd.DataFrame(data)
df
```

```python
#1 bar graph of product and shipping quantity
quantity=df.groupby('Product_name')['Weight(kg)'].sum().reset_index()
plt.figure(figsize=(10,6))
sns.barplot(x='Product_name',y='Weight(kg)',data=quantity)
plt.title('Product shipping quantity')
plt.xlabel('Product_name')
plt.ylabel('Quantity shipped')
plt.show()
```

```python
#2-Lineplot of shipping cost over shipping date
df['Month']=df['Shipping_date'].dt.month
monthly_shipping=df.groupby('Month')['Shipping_cost'].sum().reset_index()
plt.figure(figsize=(8,6))
sns.lineplot(x='Month',y='Shipping_cost',data=monthly_shipping)
plt.title('Shipping cost over Shipping date')
plt.xlabel('Month')
plt.ylabel('Shipping cost')
plt.xticks(range(1,13))
plt.grid()
plt.show()
```

```python
#3-Piechart of shipping status
shipping_status=df['Shipping_status'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(shipping_status,labels=shipping_status.index,autopct='%1.1f%%',startangle=90)
plt.title('Shipping Status distribution')
plt.show()
```

```python
#4-Boxplot of weight by destination
plt.figure(figsize=(8,6))
sns.boxplot(x=df['Destination'],y=df['Weight(kg)'])
plt.title('Weight by destination')
plt.xlabel('Destination')
plt.ylabel('Weight')
plt.xticks(rotation=45)
plt.show()
```

```python
#5-Histogram of delivery time
plt.figure(figsize=(8,6))
sns.histplot(df['Delivery_time'],bins=10,kde=True) #kde gets a waving line on graph
plt.title('Distribution of delivery days')
plt.xlabel('Delivery days')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```

```python
#6-Scatter plot of shipping cost vs date
plt.figure(figsize=(12,6))
sns.scatterplot(x='Shipping_date',y='Shipping_cost',data=df,hue='Shipping_cost')
plt.title('Shipping cost vs data')
plt.xlabel('Shipping date')
plt.ylabel('Shipping cost')
plt.legend(title='Shipping cost')
plt.show()
```

```python
#7-Graph of priority shipping by product by own choice

plt.figure(figsize=(14,6))
sns.barplot(data=df,x='Product_name',y='Priority_shipping')
plt.title('Priority shipping by product name')
plt.xlabel('Priority name')
plt.ylabel('Priority shipping count(True)')
plt.show()
```

```python
#8-line plot of shipping cost over time by destination
df['Month']=df['Shipping_date'].dt.month
monthly_shipping=df.groupby('Month')['Shipping_cost'].sum().reset_index()
plt.figure(figsize=(8,6))
sns.lineplot(x='Month',y='Shipping_cost',data=monthly_shipping)
plt.title('Shipping cost over Shipping date')
plt.xlabel('Month')
plt.ylabel('Shipping cost')
plt.xticks(range(1,13))
plt.grid()
plt.show()
```
