```python
#Create a basic plot
import matplotlib.pyplot as plt
rollno = [10,20,30,40,50]
marks = [45,30,35,48,29]

plt.plot(rollno,marks)
plt.xlabel('Roll_No')
plt.ylabel('Marks')
plt.title('Simple plot')
plt.plot()
```

```python
rollno = [10,20,30,40,50]
marks = [45,30,35,48,29]

plt.plot(rollno,marks, marker='o',linestyle='-',color='purple') #styling marker o creates dots
plt.xlabel('Roll_No')
plt.ylabel('Marks')
plt.title('Simple plot')
plt.grid(True)  #Adds a grid
plt.plot()
```

```python
rollno = [10,20,30,40,50]
marks = [45,30,35,48,29]

plt.scatter(rollno,marks, marker='x',color='r',label='Data points')
plt.xlabel('Roll_No')
plt.ylabel('Marks')
plt.title('Scatter plot')
plt.legend()
plt.grid(True)  #Adds a grid
plt.plot()
```

```python
name= ['Avani','Gauri','Deepali','Priya','Anjali']
marks = [45,30,35,48,29]

plt.bar(name,marks,color='g')
plt.xlabel('name',color='b')
plt.ylabel('Marks')
plt.title('Bar plot')
plt.plot()
```

```python
import numpy as np

data = np.random.randn(1000)  #Generate random data

plt.hist(data,bins=40,color='orange',alpha=0.8) #alpha changes the transperency of color, and bins will make it more dense
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')

plt.plot()
```

```python
data = np.random.randn(100)  #Generate random data

plt.boxplot(data)
plt.xlabel('Data')
plt.title('Box plot')
plt.show()
```

```python
data = np.random.randn(100)  #Generate random data

plt.violinplot(data,showmedians=True)
plt.xlabel('Data')
plt.title('Violin plot')
plt.show()
```

```python
rollno = [10,20,30,40,50]
marks = [45,30,35,48,29]

plt.plot(rollno,marks,color='magenta',linestyle='--',marker='s')
plt.xlabel('Roll_No')
plt.ylabel('Marks')
plt.title('Customised line plot')
plt.plot()
```

```python
rollno = [10,20,30,40,50]
marks = [45,30,35,48,29]

plt.plot(rollno,marks,color='cyan',linestyle=':',marker='o')
plt.xlabel('Roll_No')
plt.ylabel('Marks')
plt.title('Customised line plot')
plt.plot()
```

```python
rollno = [10,20,30,40,50]
marks = [45,30,35,48,29]

plt.plot(rollno,marks,color='cyan',linestyle='dashdot',marker='s')
plt.xlabel('Roll_No')
plt.ylabel('Marks')
plt.title('Customised line plot')
plt.text(1,1,'Important Point',fontsize=12,color='red')

plt.show()
```

```python
rollno = [1,2,3,4,5]
marks = [4,6,10,14,16]

plt.plot(rollno,marks,color='cyan',linestyle='dashdot',marker='s')
plt.xlabel('Roll_No')
plt.ylabel('Marks')
plt.title('Customised line plot')
plt.annotate('Annotated Point',xy=(3,10),xytext=(4,8),arrowprops=dict(arrowstyle='->',color='green'),fontsize=12,color='green')

plt.show()
```
