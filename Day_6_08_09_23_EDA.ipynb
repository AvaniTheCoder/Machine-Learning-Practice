```python
#Exploratory data analysis

import pandas as pd
import random

#create a random seed for reproducibility
random.seed(42)

#Generate synthetic data for dataframe

data = {
    'Date':pd.date_range(start='2023-01-01',periods=20,freq='D'),
    'StoreID':random.choices(range(1,6),k=20),
    'Sales':[random.randint(5000,15000) for _ in range(20)],
    'Footfall':[random.randint(200,500) for _ in range(20)],
    'Avg_Spend_Per_Customer':[random.uniform(10,50) for _ in range(20)],
    'Promotion_Type':random.choices(['Sale','Discount','None'],k=20),
    'Season':random.choices(['Spring','Summer','Fall','Winter'],k=20),
    'Weather':random.choices(['Sunny','Rainy','Cloudy'],k=20),
    'Customer_Age':[random.randint(18,65) for _ in range(20)],
    'Customer_Gender':random.choices(['Male','Female'],k=20)
}

mall_df = pd.DataFrame(data)
mall_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#set up the figure and subplots
fig, axes = plt.subplots(3,3,figsize=(15,12))

#Plot 1: Sales over Time
sns.lineplot(x='Date',y='Sales',data=mall_df,ax=axes[0,0])
axes[0,0].set_title('Sales Over time')

#Plot 2: Footfall over Time
sns.lineplot(x='Date',y='Footfall',data=mall_df,ax=axes[0,1])
axes[0,1].set_title('Footfall Over time')

#Plot 3: Average spend per customer Distribution
sns.histplot(mall_df['Avg_Spend_Per_Customer'],kde=True,ax=axes[0,2])
axes[0,2].set_title('Average spend per customer Distribution')

#Plot 4:Sales by promotion type
sns.barplot(x='Promotion_Type',y='Sales',data=mall_df,ax=axes[1,0])
axes[1,0].set_title('Sales by promotion type')

#Plot 5:Footfall by weather
sns.boxplot(x='Weather',y='Footfall',data=mall_df,ax=axes[1,1])
axes[1,1].set_title('Footfall by weather')

#Plot 6:Average spend per customer by season
sns.boxplot(x='Season',y='Avg_Spend_Per_Customer',data=mall_df,ax=axes[1,2])
axes[1,2].set_title('Average spend per customer by season')

#Plot 7:Gender distribution of customers
sns.countplot(x='Customer_Gender',data=mall_df,ax=axes[2,0])
axes[2,0].set_title('Gender distribution of customers')

#Remove the empty subplot
fig.delaxes(axes[2,1])

#Adjust layout
plt.tight_layout()

#Show the plots
plt.show()
```

```python
#Create an agricultural dataframe with the following columns,
#1-Month,2-Farm_id,3-Total production(tons),4-Amount of water consume(ltr),
#5-Avg price of per kg production(float),6-Crop type,7-Weather,8-Season,9-Temperature

import pandas as pd
import random

data={
    'Month':pd.date_range(start='2023-01-01',periods=10,freq='M'),
    'Farm_id':random.choices(range(1,6),k=10),
    'Total_production':[50,100,150,80,100,50,120,200,140,70],
    'Amount_of_water':[100,200,300,160,200,100,240,400,280,140],
    'Avg_price_per_kg_production':[3,1.5,1,2.8,3,2,0.5,1.8,2.8,1.7],
    'Crop_type':random.choices(['Rice','Wheat','Corn','Soyabean'],k=10),
    'Weather':random.choices(['Sunny','Rainy','Cloudy'],k=10),
    'Season':random.choices(['Spring','Summer','Fall','Winter'],k=10),
    'Temperature':[25.2,29.4,32,26.1,30,25,34,36,29.7,27.4]
}

agri_df=pd.DataFrame(data)
agri_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#set up the figure and subplots
fig, axes = plt.subplots(3,3,figsize=(15,12))

#Plot 1: Production over time
sns.lineplot(x='Month',y='Total_production',data=agri_df,ax=axes[0,0])
axes[0,0].set_title('Production Over time')



#Plot 3: Average spend per customer Distribution
sns.histplot(mall_df['Avg_Spend_Per_Customer'],kde=True,ax=axes[0,2])
axes[0,2].set_title('Average spend per customer Distribution')

#Plot 4:Total Production based on amount of water
sns.barplot(x='Amount_of_water',y='Total_production',data=agri_df,ax=axes[1,0])
axes[1,0].set_title('Total Production based on amount of water')

#Plot 5:Total Production by weather
sns.boxplot(x='Weather',y='Total_production',data=agri_df,ax=axes[1,1])
axes[1,1].set_title('Total Production by weather')

#Plot 6:Average spend per customer by season
sns.boxplot(x='Season',y='Avg_price_per_kg_production',data=agri_df,ax=axes[1,2])
axes[1,2].set_title('Average spend per customer by season')

#Plot 7:Crop Type distribution
sns.countplot(x='Crop_type',data=agri_df,ax=axes[2,0])
axes[2,0].set_title('Crop Type distribution')

#Remove the empty subplot
fig.delaxes(axes[2,1])

#Adjust layout
plt.tight_layout()

#Show the plots
plt.show()
```
